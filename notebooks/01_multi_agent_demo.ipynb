# Autonomous Multi-Agent Research System
## Demo Notebook
This notebook demonstrates the multi-agent research pipeline with LangGraph.

```python
import os
from dotenv import load_dotenv
load_dotenv()

from graph.research_graph import build_research_graph

graph = build_research_graph()
print('Graph compiled successfully')
```

```python
import time
import uuid

query = 'Latest developments in LLM evaluation and hallucination detection 2024'
session_id = str(uuid.uuid4())

initial_state = {
    'query': query,
    'subtasks': [],
    'search_results': [],
    'retrieved_docs': [],
    'critic_feedback': None,
    'critic_passed': False,
    'retry_count': 0,
    'synthesized_findings': '',
    'final_report': '',
    'citations': [],
    'confidence_score': 0.0,
    'current_step': 'plan',
    'session_id': session_id
}

config = {'configurable': {'thread_id': session_id}}
start = time.time()
result = graph.invoke(initial_state, config)
elapsed = time.time() - start

print(f'Research completed in {elapsed:.1f}s')
print(f'Confidence Score: {result["confidence_score"]:.2f}')
print(f'Citations: {len(result["citations"])}')
print('\n--- REPORT ---\n')
print(result['final_report'])
```
